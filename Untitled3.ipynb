```python
import numpy as np

# --- THE MONKEY PATCH ---
# We intercept numpy's clip function to prevent the MNE float/integer crash
_original_clip = np.clip

def _patched_clip(a, a_min, a_max, out=None, **kwargs):
    # If MNE tries to clip an integer array using infinity, swap infinity for the max safe integer
    if a_max == np.inf and out is not None and np.issubdtype(out.dtype, np.integer):
        a_max = np.iinfo(out.dtype).max
    return _original_clip(a, a_min, a_max, out=out, **kwargs)

# Apply the patch
np.clip = _patched_clip
# ------------------------

import mne

def preprocess_eeg_data(file_path):
    print(f"Loading {file_path}...")
    # Load the raw GDF file
    raw = mne.io.read_raw_gdf(file_path, preload=True, verbose='ERROR')

    # Select only the specific EEG channels used in the paper
    eeg_channels = ['EEG:C3', 'EEG:Cz', 'EEG:C4']
    raw.pick_channels([ch for ch in raw.ch_names if any(eeg in ch for eeg in eeg_channels)])

    # Apply Band-pass filter (0.5 - 100 Hz) and Notch (50 Hz)
    print("Applying filters (0.5-100Hz Bandpass, 50Hz Notch)...")
    raw.filter(l_freq=0.5, h_freq=100.0, fir_design='firwin', verbose='ERROR')
    raw.notch_filter(freqs=50.0, verbose='ERROR')

    # Extract Events dynamically
    events, event_dict = mne.events_from_annotations(raw, verbose='ERROR')

    # The BCI dataset annotations are strings '769' (Left) and '770' (Right)
    left_id = event_dict.get('769')
    right_id = event_dict.get('770')

    if left_id is None or right_id is None:
        raise ValueError(f"Could not find MI events 769/770 in the file. Found: {event_dict}")

    custom_event_id = {'Left_MI': left_id, 'Right_MI': right_id}

    # Segment the data into 4-second trials (1000 data points)
    tmin = 0.5
    tmax = 4.5 - (1/250)

    print("Segmenting data into 4-second trials...")
    epochs = mne.Epochs(raw, events, custom_event_id, tmin, tmax, baseline=None, preload=True, verbose='ERROR')

    # Extract the numerical array
    X = epochs.get_data()

    # Create clean binary labels: 0 for Left, 1 for Right
    y = epochs.events[:, -1]
    y = np.where(y == left_id, 0, 1)

    print(f"Preprocessing complete! Data shape: {X.shape}")
    return X, y

# Run the extraction
file_name = '/content/B0901T.gdf'
X_train, y_train = preprocess_eeg_data(file_name)
```

```python
import numpy as np
from scipy.signal import butter, filtfilt

def butter_bandpass(lowcut, highcut, fs, order=4):
    """Creates a Butterworth bandpass filter."""
    nyq = 0.5 * fs # Nyquist frequency
    low = lowcut / nyq
    high = highcut / nyq
    # Get filter coefficients
    b, a = butter(order, [low, high], btype='band')
    return b, a

def apply_filter_bank(X, fs=250.0):
    """
    Applies the filter bank to the EEG data.
    X shape expected: (trials, channels, timepoints)
    """
    print("Applying filter bank to separate frequency bands...")

    # Define the frequency bands as specified in the paper
    bands = {
        'delta': (1.0, 4.0),
        'theta': (4.0, 8.0),
        'alpha': (8.0, 13.0),
        'beta':  (13.0, 30.0)
    }

    # Initialize a dictionary to hold the filtered data for each band
    filtered_data = {}

    for band_name, (low, high) in bands.items():
        # Get the filter coefficients for this specific band
        b, a = butter_bandpass(low, high, fs)

        # Apply the zero-phase filter to the data (along the time axis = -1)
        # filtfilt prevents phase shifting which is critical for EEG temporal features
        X_filtered = filtfilt(b, a, X, axis=-1)

        filtered_data[band_name] = X_filtered
        print(f"  -> Extracted {band_name}-band ({low}-{high} Hz): Shape {X_filtered.shape}")

    return filtered_data

# --- Execute the Filter Bank ---
# Pass the X_train matrix we generated in Step 2
X_train_bands = apply_filter_bank(X_train, fs=250.0)

# Example: To access just the alpha band data later:
# alpha_data = X_train_bands['alpha']
```

```python
import numpy as np
import scipy.signal as signal
from fooof import FOOOFGroup

def extract_de_features(X_bands):
    """Calculates Differential Entropy for each channel and frequency band."""
    print("Extracting Differential Entropy (DE) features...")
    bands = ['delta', 'theta', 'alpha', 'beta']
    trials, channels, _ = X_bands['delta'].shape

    # Initialize array to hold DE features: shape (trials, channels, 4 bands)
    de_matrix = np.zeros((trials, channels, len(bands)))

    for b_idx, band in enumerate(bands):
        data = X_bands[band]
        # Calculate variance along the time axis
        variance = np.var(data, axis=-1)
        # Apply the DE formula: 0.5 * log(2 * pi * e * variance)
        # Added a tiny epsilon (1e-9) to prevent log(0) errors
        de = 0.5 * np.log(2 * np.pi * np.e * (variance + 1e-9))
        de_matrix[:, :, b_idx] = de

    # Flatten the features per trial so it can be fed into the Dense layer later
    # New shape: (trials, channels * bands) -> (120, 12)
    return de_matrix.reshape(trials, -1)

def extract_nps_features(X, fs=250.0):
    """Extracts aperiodic Neural Power Spectra features using FOOOF."""
    print("Extracting Neural Power Spectra (NPS) features...")
    trials, channels, _ = X.shape

    # 1. Calculate the Power Spectral Density (PSD) using Welch's method
    # We use a 2-second window (fs*2) to get good frequency resolution
    freqs, spectra = signal.welch(X, fs=fs, nperseg=int(fs*2))

    # Reshape spectra to process all trials and channels at once in FOOOFGroup
    # Shape becomes (trials * channels, frequencies)
    spectra_reshaped = spectra.reshape(-1, spectra.shape[-1])

    # 2. Initialize FOOOF parameterizing model (1 to 30 Hz range)
    fg = FOOOFGroup(peak_width_limits=[1.0, 8.0], max_n_peaks=4, verbose=False)
    fg.fit(freqs, spectra_reshaped, [1.0, 30.0])

    # 3. Extract the aperiodic parameters (Offset and Exponent)
    # The paper uses these to account for baseline brain state shifts
    aperiodic_params = fg.get_params('aperiodic_params')

    # Reshape back to (trials, channels * 2 features) -> (120, 6)
    nps_features = aperiodic_params.reshape(trials, -1)

    return nps_features

# --- Execute the Feature Extraction ---
# Calculate DE using the separated bands from Step 3
de_features = extract_de_features(X_train_bands)

# Calculate NPS using the broadband filtered data from Step 2
nps_features = extract_nps_features(X_train)

# Combine them into a single intrinsic feature array per trial
intrinsic_features = np.concatenate((de_features, nps_features), axis=1)

print(f"Feature Extraction Complete!")
print(f"DE Features shape: {de_features.shape}")
print(f"NPS Features shape: {nps_features.shape}")
print(f"Combined Intrinsic Features shape: {intrinsic_features.shape}")
```

```python
import numpy as np

def add_gaussian_noise(X, noise_level=0.01):
    """Adds zero-mean Gaussian noise to the data."""
    noise = np.random.normal(0, noise_level, X.shape)
    return X + noise

def segment_and_recombine(X, y, num_segments=4):
    """
    Splits trials into chunks and recombines them with other trials of the SAME class.
    Paper uses 4 segments of 250 data points (1s each).
    """
    trials, channels, timepoints = X.shape
    segment_length = timepoints // num_segments

    X_new = np.zeros_like(X)

    # We do this class by class (Left=0, Right=1)
    for class_label in np.unique(y):
        # Find all trial indices belonging to this class
        class_indices = np.where(y == class_label)[0]

        # We'll create exactly one new recombinant trial for each original trial
        for i, target_idx in enumerate(class_indices):
            recombined_trial = np.zeros((channels, timepoints))

            # For each segment, randomly pick a donor trial from the same class
            for seg in range(num_segments):
                donor_idx = np.random.choice(class_indices)
                start = seg * segment_length
                end = start + segment_length

                # Copy the segment from the donor to the new trial
                recombined_trial[:, start:end] = X[donor_idx, :, start:end]

            # Place the new recombinant trial in the array
            X_new[target_idx] = recombined_trial

    return X_new

def window_slicing(X, slice_ratio=0.9):
    """Takes a random 90% slice and linearly interpolates it back to original length."""
    trials, channels, timepoints = X.shape
    slice_length = int(timepoints * slice_ratio)

    X_new = np.zeros_like(X)

    for i in range(trials):
        # Pick a random starting point for the 90% slice
        start_idx = np.random.randint(0, timepoints - slice_length)
        end_idx = start_idx + slice_length
        sliced_data = X[i, :, start_idx:end_idx]

        # Interpolate each channel back to the original 1000 points
        for c in range(channels):
            X_new[i, c, :] = np.interp(
                np.linspace(0, 1, timepoints),
                np.linspace(0, 1, slice_length),
                sliced_data[c, :]
            )
    return X_new

def window_warping(X, warp_ratio=0.1, warp_speed=2.0):
    """Warps a random 10% window by speeding it up or slowing it down."""
    trials, channels, timepoints = X.shape
    window_length = int(timepoints * warp_ratio)

    # We need to pick a speed: either speed up (2.0) or slow down (0.5)
    speed_factor = warp_speed if np.random.rand() > 0.5 else 1.0 / warp_speed
    warped_length = int(window_length / speed_factor)

    X_new = np.zeros_like(X)

    for i in range(trials):
        start_idx = np.random.randint(0, timepoints - window_length)
        end_idx = start_idx + window_length

        for c in range(channels):
            # Extract the middle window
            window_data = X[i, c, start_idx:end_idx]

            # Interpolate the window to warp its speed
            warped_window = np.interp(
                np.linspace(0, 1, warped_length),
                np.linspace(0, 1, window_length),
                window_data
            )

            # Piece it all back together
            part1 = X[i, c, :start_idx]
            part2 = warped_window
            part3 = X[i, c, end_idx:]

            combined = np.concatenate([part1, part2, part3])

            # Because the length changed, we need to resample the entire thing
            # back to exactly 1000 data points to fit the CNN input
            X_new[i, c, :] = np.interp(
                np.linspace(0, 1, timepoints),
                np.linspace(0, 1, len(combined)),
                combined
            )

    return X_new

def apply_all_augmentations(X, y):
    """
    Applies all 4 DA methods and concatenates them with the original data.
    """
    print("Applying Data Augmentations to expand dataset...")

    # 1. Gaussian Noise (using standard deviation 0.01 as an example)
    X_gn = add_gaussian_noise(X, noise_level=0.01)

    # 2. Signal Segmentation & Recombination
    X_sr = segment_and_recombine(X, y)

    # 3. Window Slicing
    X_ws = window_slicing(X)

    # 4. Window Warping
    X_ww = window_warping(X)

    # Combine everything together along the trials axis (axis=0)
    X_augmented = np.concatenate([X, X_gn, X_sr, X_ws, X_ww], axis=0)

    # Because we made 4 copies of the data, we need 4 copies of the labels
    y_augmented = np.concatenate([y, y, y, y, y], axis=0)

    print(f"Original Data Shape: {X.shape}")
    print(f"Augmented Data Shape: {X_augmented.shape}")

    return X_augmented, y_augmented

# --- Execute Data Augmentation ---
# We augment the broadband X_train matrix before we pass it through the Filter Bank
X_train_aug, y_train_aug = apply_all_augmentations(X_train, y_train)

# NOTE: Because we augmented the raw data, we now need to re-run the Filter Bank
# and Feature Extraction on this NEW, LARGER dataset so they match up!
print("\nRe-running Filter Bank and Feature Extraction on augmented data...")
X_train_bands_aug = apply_filter_bank(X_train_aug)
de_features_aug = extract_de_features(X_train_bands_aug)
nps_features_aug = extract_nps_features(X_train_aug)
intrinsic_features_aug = np.concatenate((de_features_aug, nps_features_aug), axis=1)

print("\n--- Final Training Tensors Ready! ---")
print(f"Augmented EEG Bands: {X_train_bands_aug['alpha'].shape}")
print(f"Augmented Intrinsic Features: {intrinsic_features_aug.shape}")
print(f"Augmented Labels: {y_train_aug.shape}")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, Input

def create_mscb(input_tensor, block_name):
    """
    Creates a Multi-Scale Convolution Block (MSCB) as defined in the paper.
    Uses kernels of 1x5 (Large), 1x3 (Medium), and 1x1 (Small).
    """
    # The paper uses a uniform pooling stride of 5 so all branches
    # output the same dimensions and can be concatenated.
    pool_stride = (1, 5)

    # Branch 1: Large Kernel
    b1 = layers.Conv2D(14, (1, 5), padding='same', activation='relu', name=f'{block_name}_L_conv')(input_tensor)
    b1 = layers.MaxPooling2D((1, 5), strides=pool_stride, padding='same', name=f'{block_name}_L_pool')(b1)

    # Branch 2: Medium Kernel
    b2 = layers.Conv2D(14, (1, 3), padding='same', activation='relu', name=f'{block_name}_M_conv')(input_tensor)
    b2 = layers.MaxPooling2D((1, 3), strides=pool_stride, padding='same', name=f'{block_name}_M_pool')(b2)

    # Branch 3: Small Kernel
    b3 = layers.Conv2D(14, (1, 1), padding='same', activation='relu', name=f'{block_name}_S_conv')(input_tensor)
    b3 = layers.MaxPooling2D((1, 1), strides=pool_stride, padding='same', name=f'{block_name}_S_pool')(b3)

    # Branch 4: Max Pooling + Conv
    b4 = layers.MaxPooling2D((1, 3), strides=pool_stride, padding='same', name=f'{block_name}_P_pool')(input_tensor)
    b4 = layers.Conv2D(24, (1, 3), padding='same', activation='relu', name=f'{block_name}_P_conv')(b4)

    # Concatenate all 4 branches along the feature map axis (axis=-1)
    concat = layers.Concatenate(axis=-1, name=f'{block_name}_concat')([b1, b2, b3, b4])

    # Followed by the shared convolution and pooling layer for the block
    x = layers.Conv2D(112, (1, 3), strides=(1, 1), padding='same', activation='relu', name=f'{block_name}_out_conv')(concat)
    x = layers.MaxPooling2D((1, 5), strides=(1, 5), padding='same', name=f'{block_name}_out_pool')(x)

    # Flatten to 1D for the Dense layers
    return layers.Flatten(name=f'{block_name}_flatten')(x)

def build_ms_cnn(channels=3, timepoints=1000, num_intrinsic_features=18):
    """Builds the complete Multi-Input CNN."""

    # 1. Define the 4 inputs for the frequency bands
    # Shape: (channels, timepoints, 1) because Conv2D expects a channel dimension at the end
    input_shape = (channels, timepoints, 1)

    in_delta = Input(shape=input_shape, name='input_delta')
    in_theta = Input(shape=input_shape, name='input_theta')
    in_alpha = Input(shape=input_shape, name='input_alpha')
    in_beta  = Input(shape=input_shape, name='input_beta')

    # 2. Define the input for the intrinsic features
    in_intrinsic = Input(shape=(num_intrinsic_features,), name='input_intrinsic')

    # 3. Pass each frequency band through its own MSCB
    mscb_delta = create_mscb(in_delta, 'MSCB_Delta')
    mscb_theta = create_mscb(in_theta, 'MSCB_Theta')
    mscb_alpha = create_mscb(in_alpha, 'MSCB_Alpha')
    mscb_beta  = create_mscb(in_beta, 'MSCB_Beta')

    # 4. Feature Fusion: Combine the flattened CNN outputs and the Intrinsic Features
    fused_features = layers.Concatenate(name='Feature_Fusion')([
        mscb_delta, mscb_theta, mscb_alpha, mscb_beta, in_intrinsic
    ])

    # 5. Fully Connected (FC) Layers
    fc1 = layers.Dense(400, activation='relu', name='FC_1')(fused_features)
    fc1 = layers.Dropout(0.5, name='Dropout')(fc1) # Prevents overfitting

    fc2 = layers.Dense(300, activation='relu', name='FC_2')(fc1)

    # 6. Output Layer: 2 neurons for Left MI and Right MI with Softmax
    output = layers.Dense(2, activation='softmax', name='Output_Layer')(fc2)

    # Compile the final model
    model = models.Model(
        inputs=[in_delta, in_theta, in_alpha, in_beta, in_intrinsic],
        outputs=output,
        name="MS_CNN_BCI"
    )

    return model

# --- Instantiate the Model ---
model = build_ms_cnn()

# Print the architecture summary
model.summary()
```

```python
import tensorflow as tf

def prepare_tensors_for_training(X_bands, intrinsic_features, y):
    """
    Expands the dimensions of the input arrays so they match the
    (batch_size, channels, timepoints, 1) shape expected by Conv2D.
    """
    print("Formatting arrays for Keras...")

    # Add a dummy axis at the end of each frequency band array
    X_delta = np.expand_dims(X_bands['delta'], axis=-1)
    X_theta = np.expand_dims(X_bands['theta'], axis=-1)
    X_alpha = np.expand_dims(X_bands['alpha'], axis=-1)
    X_beta  = np.expand_dims(X_bands['beta'], axis=-1)

    return [X_delta, X_theta, X_alpha, X_beta, intrinsic_features], y

# --- 1. Prepare Data ---
# Use the augmented data we generated in Step 5
X_train_inputs, y_train_final = prepare_tensors_for_training(
    X_train_bands_aug,
    intrinsic_features_aug,
    y_train_aug
)

# --- 2. Compile the Model ---
# The paper uses SGD with a specific learning rate
optimizer = tf.keras.optimizers.SGD(learning_rate=0.001)

model.compile(
    optimizer=optimizer,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# --- 3. Set up Learning Rate Decay ---
# The paper mentions the learning rate decays every 30 epochs with a rate of 0.5
def step_decay(epoch):
    initial_lrate = 0.001
    drop = 0.5
    epochs_drop = 30.0
    # Calculate the new learning rate based on the epoch number
    lrate = initial_lrate * (drop ** np.floor((1+epoch)/epochs_drop))
    return lrate

lr_scheduler = tf.keras.callbacks.LearningRateScheduler(step_decay)

# --- 4. Train the Model! ---
print("\nInitiating Training Phase...")
print("Note: The paper trains for 500 epochs. We are running 50 here as a quick test.")

# We use 50 epochs here just so you don't have to wait an hour on Colab,
# but you can change this to 500 to match the paper exactly.
history = model.fit(
    x=X_train_inputs,
    y=y_train_final,
    batch_size=16,          # Paper specifies batch size of 16
    epochs=50,              # Change to 500 for full replication
    validation_split=0.2,   # Use 20% of augmented data to monitor validation accuracy
    callbacks=[lr_scheduler],
    verbose=1
)

print("\nTraining Complete!")
```
